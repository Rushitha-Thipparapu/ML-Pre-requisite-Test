For finding the elements of C which lies in row i and column j, Combine the i-th row of A and the j-th column of B together. Then multiply each of Cij parts and add all products to get the final value of Cij.
As there are k columns in A and k rows in B, you multiply k pairs of elements for each element in C. And n×m elements are present in the resulting matrix
C, n×m×k multiplications need to be done.
As k numbers are added, k-1 additions are needed. So,n×m×(k-1) additions are needed.

7. Create a Python nested dictionary to represent hierarchical information. We want to store record of students and their marks. Something like:

    1. 2022
       1. Branch 1
       1. Roll Number: 1, Name: N, Marks:
           1. Maths: 100, English: 70 …
    2. Branch 2
       2. 2023

        Branch 1
        Branch 2
    3. 2024
        Branch 1
        Branch 2
    4. 2025
        Branch 1
        Branch 2

```python
class Student:
    def __init__(self, id, name, grades):
        self.id = id
        self.name = name
        self.grades = grades

    def __repr__(self):
        return f"Student(ID={self.id}, Name='{self.name}', Grades={self.grades})"

class Department:
    def __init__(self, name):
        self.name = name
        self.students = {}

    def add_student(self, student):
        self.students[student.id] = student

    def __repr__(self):
        return f"Department(Name='{self.name}', Students={self.students})"

class AcademicYear:
    def __init__(self, year):
        self.year = year
        self.departments = {}

    def add_department(self, department):
        self.departments[department.name] = department

    def __repr__(self):
        return f"AcademicYear(Year={self.year}, Departments={self.departments})"

students_2022_branch_1 = [
    Student(101, "John Doe", {"Maths": 85, "English": 90}),
    Student(102, "Alice Johnson", {"Maths": 78, "English": 82})
]
students_2022_branch_2 = [
    Student(103, "Jane Smith", {"Maths": 80, "English": 88}),
    Student(104, "Bob Brown", {"Maths": 76, "English": 84})
]

students_2023_branch_1 = [
    Student(105, "Emily Davis", {"Maths": 92, "English": 95}),
    Student(106, "Michael Clark", {"Maths": 87, "English": 90})
]
students_2023_branch_2 = [
    Student(107, "Jessica Taylor", {"Maths": 88, "English": 85}),
    Student(108, "David Wilson", {"Maths": 82, "English": 78})
]

students_2024_branch_1 = [
    Student(109, "Sarah Miller", {"Maths": 91, "English": 89}),
    Student(110, "James Lee", {"Maths": 84, "English": 86})
]
students_2024_branch_2 = [
    Student(111, "Laura Davis", {"Maths": 77, "English": 80}),
    Student(112, "Daniel Smith", {"Maths": 83, "English": 74})
]

students_2025_branch_1 = [
    Student(113, "Sophia Brown", {"Maths": 79, "English": 91}),
    Student(114, "Liam Johnson", {"Maths": 82, "English": 89})
]
students_2025_branch_2 = [
    Student(115, "Olivia Jones", {"Maths": 85, "English": 90}),
    Student(116, "Noah Williams", {"Maths": 78, "English": 77})
]

def create_year_with_departments(year, branch_1_students, branch_2_students):
    year_obj = AcademicYear(year)

    branch_1 = Department("Branch 1")
    for student in branch_1_students:
        branch_1.add_student(student)

    branch_2 = Department("Branch 2")
    for student in branch_2_students:
        branch_2.add_student(student)

    year_obj.add_department(branch_1)
    year_obj.add_department(branch_2)

    return year_obj

year_2022 = create_year_with_departments(2022, students_2022_branch_1, students_2022_branch_2)
year_2023 = create_year_with_departments(2023, students_2023_branch_1, students_2023_branch_2)
year_2024 = create_year_with_departments(2024, students_2024_branch_1, students_2024_branch_2)
year_2025 = create_year_with_departments(2025, students_2025_branch_1, students_2025_branch_2)

records = {
    2022: year_2022,
    2023: year_2023,
    2024: year_2024,
    2025: year_2025
}

for year, year_obj in records.items():
    print(f"Academic Year: {year_obj.year}")
    for dept_name, dept_obj in year_obj.departments.items():
        print(f"  Department: {dept_obj.name}")
        for student in dept_obj.students.values():
            print(f"    {student}")
    print()
```

8. Store the same information using Python classes. We have an overall database which is a list of year objects. Each year contains a list of branches. Each branch contains a list of students. Each student has some properties like name, roll number and has marks in some subjects.

```python
class Student:
    def __init__(self, roll_number, name, marks):
        self.roll_number = roll_number
        self.name = name
        self.marks = marks

    def __repr__(self):
        return (f"Student(Roll Number={self.roll_number}, Name='{self.name}', "
                f"Marks={self.marks})")

class Branch:
    def __init__(self, name):
        self.name = name
        self.students = []

    def add_student(self, student):
        self.students.append(student)

    def __repr__(self):
        return f"Branch(Name='{self.name}', Students={self.students})"

class AcademicYear:
    def __init__(self, year):
        self.year = year
        self.branches = []

    def add_branch(self, branch):
        self.branches.append(branch)

    def __repr__(self):
        return f"AcademicYear(Year={self.year}, Branches={self.branches})"

class Database:
    def __init__(self):
        self.years = []

    def add_year(self, year):
        self.years.append(year)

    def __repr__(self):
        return f"Database(Years={self.years})"

students_data = {
    2022: {
        "Branch 1": [
            Student(101, "John Doe", {"Maths": 85, "English": 90}),
            Student(102, "Alice Johnson", {"Maths": 78, "English": 82}),
        ],
        "Branch 2": [
            Student(103, "Jane Smith", {"Maths": 80, "English": 88}),
            Student(104, "Bob Brown", {"Maths": 76, "English": 84}),
        ],
    },
    2023: {
        "Branch 1": [
            Student(105, "Emily Davis", {"Maths": 92, "English": 95}),
            Student(106, "Michael Clark", {"Maths": 87, "English": 90}),
        ],
        "Branch 2": [
            Student(107, "Jessica Taylor", {"Maths": 88, "English": 85}),
            Student(108, "David Wilson", {"Maths": 82, "English": 78}),
        ],
    },
    2024: {
        "Branch 1": [
            Student(109, "Sarah Miller", {"Maths": 91, "English": 89}),
            Student(110, "James Lee", {"Maths": 84, "English": 86}),
        ],
        "Branch 2": [
            Student(111, "Laura Davis", {"Maths": 77, "English": 80}),
            Student(112, "Daniel Smith", {"Maths": 83, "English": 74}),
        ],
    },
    2025: {
        "Branch 1": [
            Student(113, "Sophia Brown", {"Maths": 79, "English": 91}),
            Student(114, "Liam Johnson", {"Maths": 82, "English": 89}),
        ],
        "Branch 2": [
            Student(115, "Olivia Jones", {"Maths": 85, "English": 90}),
            Student(116, "Noah Williams", {"Maths": 78, "English": 77}),
        ],
    }
}

db = Database()

for year, branches in students_data.items():
    year_obj = AcademicYear(year)
    for branch_name, students_list in branches.items():
        branch = Branch(branch_name)
        for student in students_list:
            branch.add_student(student)
        year_obj.add_branch(branch)
    db.add_year(year_obj)

for year in db.years:
    print(f"Academic Year: {year.year}")
    for branch in year.branches:
        print(f"  Department: {branch.name}")
        for student in branch.students:
            print(f"    {student}")
    print()
```

9. Using matplotlib plot the following functions on the domain: x = 0.5 to 100.0 in steps of 0.5.

 1. y=x

 2. y=x^2

 3. y=x^3/100

 4. y=sin(x)

 5. y=sin(x)/x

 6. y=log(x)

 7. y=e^x

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.5, 100.1, 0.5)

y1 = x
y2 = x ** 2
y3 = x ** 3 / 100
y4 = np.sin(x)
y5 = np.sin(x) / x
y6 = np.log(x)
y7 = np.exp(x)

# Plot y = x
plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='y = x', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x')
plt.grid(True)
plt.legend()
plt.show()

# Plot y = x^2
plt.figure(figsize=(10, 6))
plt.plot(x, y2, label='y = x^2', color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x^2')
plt.grid(True)
plt.legend()
plt.show()

# Plot y = x^3 / 100
plt.figure(figsize=(10, 6))
plt.plot(x, y3, label='y = x^3 / 100', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x^3 / 100')
plt.grid(True)
plt.legend()
plt.show()

# Plot y = sin(x)
plt.figure(figsize=(10, 6))
plt.plot(x, y4, label='y = sin(x)', color='orange')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = sin(x)')
plt.grid(True)
plt.legend()
plt.show()

# Plot y = sin(x) / x
plt.figure(figsize=(10, 6))
plt.plot(x, y5, label='y = sin(x) / x', color='purple')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = sin(x) / x')
plt.grid(True)
plt.legend()
plt.show()

# Plot y = log(x)
plt.figure(figsize=(10, 6))
plt.plot(x, y6, label='y = log(x)', color='brown')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = log(x)')
plt.grid(True)
plt.legend()
plt.show()

# Plot y = e^x
plt.figure(figsize=(10, 6))
plt.plot(x, y7, label='y = e^x', color='cyan')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = e^x')
plt.grid(True)
plt.legend()
plt.show()
```

10. Using numpy generate a matrix of size 20X5 containing random numbers drawn uniformly from the range of 1 to 2. Using Pandas create a dataframe out of this matrix. Name the columns of the dataframe as “a”, “b”, “c”, “d”, “e”. Find the column with the highest standard deviation. Find the row with the lowest mean.

```python
import numpy as np
import pandas as pd

data_matrix = np.random.uniform(1, 2, size=(20, 5))
df = pd.DataFrame(data_matrix, columns=["a", "b", "c", "d", "e"])
std_devs = df.std()

max_col = std_devs.idxmax()
max_mean = std_devs.max()

print(f"Column with the highest standard deviation: {max_col}")
print(f"Standard deviation of this column: {max_mean}")

row_means = df.mean(axis=1)

min_mean_row_index= row_means.idxmin()
min_mean_value = row_means.min()

print(f"Row with the lowest mean: {min_mean_row_index}")
print(f"Mean of this row: {min_mean_value}")

print("\nDataFrame:")
print(df)
```

The value in the column “g” should be “LT8” if the value in the column “f” is less than 8 and “GT8” otherwise. Find the number of rows in the dataframe where the value in the column “g” is “LT8”. Find the standard deviation of the column “f” for the rows where the value in the column “g” is “LT8” and “GT8” respectively.

13. Write a function to compute the argmin of a numpy array. The function should take a numpy array as input and return the index of the minimum element. You can use the np.argmin function to verify your solution.

```python
import numpy as np

def custom_argmin(arr):
    min_value = np.min(arr)
    min_index = np.where(arr == min_value)[0][0]
    return min_index

array = np.random.randint(0, 100, size=4)
print("Array:", array)

index_custom = custom_argmin(array)
index_np = np.argmin(array)

print("Index of minimum element (custom function):", index_custom)
print("Index of minimum element (np.argmin):", index_np)
```
